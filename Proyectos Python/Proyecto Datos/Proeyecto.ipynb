```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

#Se lee el archivo como dataframe de pandas (Corriga aquí con su ruta)
amazon = pd.read_csv("D:\\Proyectos Universidad\\Proyectos-Universidad\\Proyectos Python\\Proyecto Datos\\amazon.csv")
print("Archivo leido: ")
amazon
```

```python
#Se imprimen las columnas del archivo
print("Columnas del archivo: ")
for i in amazon.columns:
    print(i)
```

```python
#Se imprimen el numero de columnas y filas del archivo
print(f"El archivo tiene {amazon.shape[0]} columnas, y {amazon.shape[1]} filas" )
```

```python
#Se muestran cuales columnas estan nulas
print("Columnas nulas:")
amazon.isnull().sum()
```

```python
#Se transforman el tipo de datos de string a numerico para facilitar ciertas operaciones
amazon['rating_count'] = pd.to_numeric(amazon['rating_count'],errors = 'coerce')
#Se corrige las entradas de datos nulas con la media de la cantidad de ratings
media_rating_count = amazon['rating_count'].mean()
amazon['rating_count'] = amazon['rating_count'].fillna(media_rating_count)
amazon.isnull().sum()
```

```python
#Se eliminan simbolos y demas cosas para facilitar operaciones dentro del codigo, ademas de volverlos tipo float
amazon['discounted_price'] = amazon['discounted_price'].str.replace("₹",'')
amazon['discounted_price'] = amazon['discounted_price'].str.replace(",",'')
amazon['discounted_price'] = amazon['discounted_price'].astype('float64')

amazon['actual_price'] = amazon['actual_price'].str.replace("₹",'')
amazon['actual_price'] = amazon['actual_price'].str.replace(",",'')
amazon['actual_price'] = amazon['actual_price'].astype('float64')
```

```python
#Se cambia el descuento, de manera que se opere como un decimal, para facilitar operaciones, tambien se cambia el tipo de dato a float
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%','').astype('float64')

amazon['discount_percentage'] = amazon['discount_percentage'] / 100
```

```python
#Se imprimen todas las ratings que aparecen, y sus cantidades respectivas
print("Ratings y sus cantidades:")
amazon['rating'].value_counts()
```

```python
#Se reemplaza por la media
amazon['rating'] = amazon['rating'].replace('|', np.nan)
amazon['rating'] = pd.to_numeric(amazon['rating'],errors='coerce')
media = amazon['rating'].mean()
amazon['rating'].fillna(media,inplace = True)
amazon['rating'].value_counts()
```

```python
#Se separan las muchas categorias de la base de datos, separandolas cada vez que se encuentre un |, ya que asi se encuentran formateadas
catsplit = amazon['category'].str.split('|', expand=True)
catsplit
```

```python
#Se separan por categorias diferentes, para entender de mejor manera
catsplit = catsplit.rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
```

```python
#Se implementan  las nuevas columnas de multiples categorias
amazon['category_1'] = catsplit['category_1']
amazon['category_2'] = catsplit['category_2']

#Se borra la columna 'category'
amazon.drop(columns='category', inplace=True)
```

```python
#Se evaluan cuantas categorias existen de cada una
amazon['category_1'].value_counts()
```

```python
#Se cambian ciertas categorias por un mejor formato, con espacios
amazon['category_1'] = amazon['category_1'].str.replace('&', ' & ')
amazon['category_1'] = amazon['category_1'].str.replace('OfficeProducts', 'Office Products')
amazon['category_1'] = amazon['category_1'].str.replace('MusicalInstruments', 'Musical Instruments')
amazon['category_1'] = amazon['category_1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
#Se evaluan la cantidad de categorias respectivamente
amazon['category_2'].value_counts()
```

```python
#Se brinda un mejor formato igualmente, con espacios mayormente
amazon['category_2'] = amazon['category_2'].str.replace('&', ' & ')
amazon['category_2'] = amazon['category_2'].str.replace(',', ', ')
amazon['category_2'] = amazon['category_2'].str.replace('HomeAppliances', 'Home Appliances')
amazon['category_2'] = amazon['category_2'].str.replace('AirQuality', 'Air Quality')
amazon['category_2'] = amazon['category_2'].str.replace('WearableTechnology', 'Wearable Technology')
amazon['category_2'] = amazon['category_2'].str.replace('NetworkingDevices', 'Networking Devices')
amazon['category_2'] = amazon['category_2'].str.replace('OfficePaperProducts', 'Office Paper Products')
amazon['category_2'] = amazon['category_2'].str.replace('ExternalDevices', 'External Devices')
amazon['category_2'] = amazon['category_2'].str.replace('DataStorage', 'Data Storage')
amazon['category_2'] = amazon['category_2'].str.replace('HomeStorage', 'Home Storage')
amazon['category_2'] = amazon['category_2'].str.replace('HomeAudio', 'Home Audio')
amazon['category_2'] = amazon['category_2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
amazon['category_2'] = amazon['category_2'].str.replace('BatteryChargers', 'Battery Chargers')
amazon['category_2'] = amazon['category_2'].str.replace('CraftMaterials', 'Craft Materials')
amazon['category_2'] = amazon['category_2'].str.replace('OfficeElectronics', 'Office Electronics')
amazon['category_2'] = amazon['category_2'].str.replace('PowerAccessories', 'Power Accessories')
amazon['category_2'] = amazon['category_2'].str.replace('CarAccessories', 'Car Accessories')
amazon['category_2'] = amazon['category_2'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
amazon['category_2'] = amazon['category_2'].str.replace('HomeTheater', 'Home Theater')
```

```python
#Se eliminan los espacios en blanco
amazon['product_id'].str.strip()
```

```python
#Se crea una nueva columna con la diferenciacion entre el precio y el descuento
amazon['difference_price'] = amazon['actual_price'] - amazon['discounted_price']
amazon.head()
```

```python
#Se crea un nuevo dataframe solo con user_id y user_name
reviewers = amazon[['user_id','user_name']]
reviewers
```

```python
#Se separa en un arreglo los user_id, user_name, review_id, review_content y review_title, para que se visualize mejor los datos, y se puedan operar mejor los datos
amazon['user_id'] = amazon['user_id'].str.split(',')
amazon['user_name'] = amazon['user_name'].str.split(',')
amazon['review_id'] = amazon['review_id'].str.split(',')
amazon['review_content'] = amazon['review_content'].str.split(',')
amazon['review_title'] = amazon['review_title'].str.split(',')
```

```python
#Se crea un nuevo arreglo que guarda solo las user_name
reviewer_name_split = reviewers['user_name'].str.split(',', expand=False)

reviewer_name_split
```

```python
#Se separa en una fila independiente cada usuario
review_name_exp = reviewer_name_split.explode()

reviewer_name_clean = review_name_exp.reset_index(drop=True)

reviewer_name_clean
```

```python
#Se crea un nuevo arreglo que guarda solo las user_id
reviewer_id_split = reviewers['user_id'].str.split(',', expand=False)

reviewer_id_split
```

```python
#Se separa en una fila independiente cada id
review_id_exp = reviewer_id_split.explode()

reviewer_id_clean = review_id_exp.reset_index(drop=True)

reviewer_id_clean
```

```python
#Se crean los dataframe respectivamente
amazon21 = pd.DataFrame(data=reviewer_id_clean)
amazon22 = pd.DataFrame(data=reviewer_name_clean)
```

```python
#Se juntan los dos dataframe
amazon2 = pd.merge(amazon21, amazon22, left_index=True, right_index=True)
amazon2
```

```python
most_main_items = amazon['category_1'].value_counts().head(5).rename_axis('category_1').reset_index(name='counts')

most_sub_items = amazon['category_2'].value_counts().head(10).rename_axis('category_2').reset_index(name='counts')

fig, ax = plt.subplots(2, 1, figsize=(8, 10))

sns.barplot(ax=ax[0], data=most_main_items, x='counts', y='category_1')
sns.barplot(ax=ax[1], data=most_sub_items, x='counts', y='category_2')

plt.subplots_adjust(hspace = 0.3)

ax[0].set_xlabel('Cantidad', fontweight='bold')
ax[0].set_ylabel('Categoria', fontweight='bold')

ax[1].set_xlabel('Cantidad', fontweight='bold')
ax[1].set_ylabel('Sub categoria', fontweight='bold')

ax[0].set_title('Mayor cantidad de productos por categoria', fontweight='bold')
ax[1].set_title('Mayor cantidad de productos por subcategoria', fontweight='bold')


ax[0].bar_label(ax[0].containers[0])
ax[1].bar_label(ax[1].containers[0])

plt.show()
```

```python
dif_price_large = sns.barplot(data= amazon.sort_values('difference_price', ascending=False).head(5), x='difference_price', y='product_name')

dif_price_large.set_title('Top 5 productos con mayor diferencia en precio (despues del descuento)', fontweight='bold')
dif_price_large.set_xlabel('Diferencia de precio', fontweight='bold')
dif_price_large.set_ylabel('Nombre del producto', fontweight='bold')

plt.show()
```
